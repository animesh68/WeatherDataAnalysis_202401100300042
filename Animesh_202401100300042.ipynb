Weather Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/weather_data.csv')
print(df.head()) # Display the first few rows of the DataFrame
```

```python
# Check the dimensions of the dataset
print(df.shape)
```

```python
# Get summary statistics of numerical variables
print(df.describe())
```

```python
# Check the data types of variables
print(df.dtypes)
```

```python
# Bar plot
sns.histplot(x='Temperature', data=df)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature')
plt.show()

# Continue with other plots, make sure column names are correct...
```

```python
# Histogram
column_to_use = 'Temperature'  # Replace with an appropriate column name
plt.hist(df[column_to_use], bins=10)
plt.xlabel(column_to_use)  # Update x-axis label accordingly
plt.ylabel('Frequency')
plt.title(f'Distribution of {column_to_use}')  # Update title accordingly
plt.show()
```

```python
import matplotlib.pyplot as plt

x_column = 'Temperature'
y_column = 'Humidity'

# Scatter plot
plt.scatter(df[x_column], df[y_column])
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'{x_column} vs. {y_column}')
plt.show()
```
